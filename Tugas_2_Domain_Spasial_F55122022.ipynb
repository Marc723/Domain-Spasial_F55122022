```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('bg4.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Difference Image
def difference_image(image1, image2):
    return cv2.absdiff(image1, image2)
# Average Filter
def average_filter(image, kernel_size=(3, 3)):
    return cv2.blur(image, kernel_size)
# Median Filter
def median_filter(image, kernel_size=3):
    return cv2.medianBlur(image, kernel_size)
# Min Filter
def min_filter(image, kernel_size=(3, 3)):
    return cv2.erode(image, np.ones(kernel_size))
# Max Filter
def max_filter(image, kernel_size=(3, 3)):
    return cv2.dilate(image, np.ones(kernel_size))
# Laplacian
def laplacian(image):
    return cv2.Laplacian(image, cv2.CV_64F)
# Gradient Sobel
def gradient_sobel(image):
    sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
    return sobel_x, sobel_y
```

```python
# Apply operations
smoothed_image = average_filter(gray_image, kernel_size=(10, 10))
median_filtered = median_filter(gray_image, kernel_size=11)
min_filtered = min_filter(gray_image, kernel_size=(3, 3))
max_filtered = max_filter(gray_image, kernel_size=(3, 3))
laplacian_image = laplacian(gray_image)
sobel_x, sobel_y = gradient_sobel(gray_image)
```

```python
# Plot results
plt.figure(figsize=(20, 20))
```

```python
fig, axes = plt.subplots(3, 3, figsize=(12, 12))

axes[0, 0].set_title("Original")
axes[0, 0].imshow(gray_image, cmap='gray')

axes[0, 1].set_title("Difference Image")
axes[0, 1].imshow(difference_image(gray_image, smoothed_image), cmap='gray')

axes[0, 2].set_title("Neighborhood Averaging")
axes[0, 2].imshow(smoothed_image, cmap='gray')

axes[1, 0].set_title("Median Filter")
axes[1, 0].imshow(median_filtered, cmap='gray')

axes[1, 1].set_title("Min Filter")
axes[1, 1].imshow(min_filtered, cmap='gray')

axes[1, 2].set_title("Max Filter")
axes[1, 2].imshow(max_filtered, cmap='gray')

axes[2, 0].set_title("Laplacian")
axes[2, 0].imshow(laplacian_image, cmap='gray')

axes[2, 1].set_title("Gradient Sobel X")
axes[2, 1].imshow(sobel_x, cmap='gray')

axes[2, 2].set_title("Gradient Sobel Y")
axes[2, 2].imshow(sobel_y, cmap='gray')

# Tampilkan gambar
plt.show()
```
